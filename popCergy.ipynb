Évolution et structure de la population : 2008-2013-2019

*Sources : Insee, RP2008, RP2013 et RP2019, exploitations principales, géographie au 01/01/2022.*


```python
import pandas as pd

data = {'2008': [56108, 12144, 17621, 12026, 9847, 3356, 1113], 
        '2013': [61708, 14169, 17085, 14117, 9739, 5203, 1395], 
        '2019': [65911,15529,17871,15362,9513,6017,1619]}

index = ['Ensemble', '0 à 14 ans', '15 à 29 ans', '30 à 44 ans', 
                 '45 à 59 ans', '60 à 74 ans', '75 ans ou plus']

dataframe = pd.DataFrame(data, index)
print(dataframe)
```

```python
# Valeurs en %
```

```python
dataframe_ratio = dataframe

dataframe_ratio['2008'] = dataframe_ratio['2008'].apply(
    lambda x: 100 * x / dataframe_ratio.loc['Ensemble','2008']).apply(
    lambda x: float("{:.2f}".format(x)))

dataframe_ratio['2013'] = dataframe_ratio['2013'].apply(
    lambda x: 100 * x / dataframe_ratio.loc['Ensemble','2013']).apply(
    lambda x: float("{:.2f}".format(x)))

dataframe_ratio['2019'] = dataframe_ratio['2019'].apply(
    lambda x: 100 * x / dataframe_ratio.loc['Ensemble','2019']).apply(
    lambda x: float("{:.2f}".format(x)))

dataframe_ratio.drop('Ensemble', inplace=True)
print(dataframe_ratio)
```

```python
import numpy as np
import matplotlib.pyplot as plt

plot_index = ['0-14 ans', '15-29 ans', '30-44 ans', 
                 '45-59 ans', '60-74 ans', '75ans +']
x_axis = np.arange(len(plot_index))

plt.bar(x_axis -0.1, dataframe_ratio['2008'], width=0.2, label = '2008')
plt.bar(x_axis +0.1, dataframe_ratio['2013'], width=0.2, label = '2013')
plt.bar(x_axis +0.3, dataframe_ratio['2019'], width=0.2, label = '2019')

plt.xticks(x_axis, plot_index)
plt.title("Population par tranches d'âge")
plt.legend()
plt.show()

```
